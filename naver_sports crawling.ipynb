```python
from bs4 import BeautifulSoup
```

```python
from selenium import webdriver
import time
```

```python
result_list = []
driver = webdriver.Chrome("./chromedriver")
for i in range(1,10) :
    driver.get("https://sports.news.naver.com/wfootball/news/index.nhn?isphoto=N&type=popular&page=" + str(i))
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    for i in soup.select("#_newsList > ul > li"):
        print(i.find("a", class_="title").text)
        result_list.append(i.find("a", class_="title").text)
    time.sleep(5)
```
