```python
# 엑셀을 읽기 위한 준비과정
from openpyxl import load_workbook

read_wb = load_workbook("./soccer.xlsx")
read_ws = read_wb.active

list_excel = []

for i in range(1, 161) :
    print(read_ws.cell(i,1).value)
    list_excel.append(read_ws.cell(i,1).value)
```

```python
list_excel
```

```python
from konlpy.tag import Kkma
import collections
```

```python
kkma = Kkma()
```

```python
list_temp = []
```

```python
for row in list_excel :
    print(kkma.nouns(row))
    list_temp = list_temp + kkma.nouns(row)
```

```python
list_temp
```

```python
result_list = []
```

```python
for check in list_temp :
#     print(check)
    if(len(check)) > 1 :
        result_list.append(check)
```

```python
result_list
```

```python
collections.Counter(result_list)
```

```python
collections.Counter(result_list).most_common(20)
```
