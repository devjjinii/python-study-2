```python
from openpyxl import load_workbook

read_wb = load_workbook('./soccer.xlsx')
read_ws = read_wb.active

list_excel = []

for i in range(1,161) :
    print(read_ws.cell(i,1).value.strip())
    list_excel.append(str(read_ws.cell(i,1).value.strip()))

print(list_excel)
```

```python
from konlpy.tag import Kkma
import collections
kkma = Kkma()

list_temp = []

for row in list_excel :
    list_temp = kkma.nouns(row) + list_temp

print(list_temp)
```

```python
list_result = []
for check in list_temp :
    if len(check) > 1:
#         print(check)
        list_result.append(check)
        
list_result

# collections.Counter(list_result)
# collections.Counter(list_result).most_common(20)

# print(collections.Counter(list_result).most_common(20))
```

```python
last_text = ""
```

```python
for i in list_result :
    last_text = last_text + " " + i
```

```python
last_text
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
wordcloud = WordCloud(font_path='C:\\Windows\\Fonts\\NanumGothicCoding.ttf').generate(last_text)
```

```python
wordcloud.words_
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```
